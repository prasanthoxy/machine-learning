```python
# iam using jupyter notebook so that i uploaded LabSession1.xlsl file to jupyter so that there is no need to set path from the computer external drive
# Calculate the mean and variance of the price data present in column D

# Reading excel file using pd.read_excel function 
# pd implies pandas module 
# Importing pandas library

import numpy as np
import pandas as pd
import statistics
from scipy.stats import norm
from matplotlib import pyplot as plt

irctc = pd.read_excel("LabSession1.xlsx", index_col=0)# specify that index_col parameter
irctc


```

```python
# converting particular column in excel to list by using tolist() function
price = irctc['Price'].tolist()
print(price)
#print(len(price))
```

```python
#import statistics
#from scipy.stats import norm
#from matplotlib import pyplot as plt

price_mean = statistics.mean(price)
print("mean of price is:", price_mean)

price_variance = statistics.variance(price)
print("Variance of price is :", price_variance)

price_stdev = statistics.stdev(price)
print("standarad deviation of price is: ", price_stdev)

# representing total length of price column vs population mean of price column
graph_mean = pd.DataFrame({
      'price': price,
      'mean': [1560.663453815261 for i in range(len(price))]})

# graph of population mean
graph_mean.plot()
plt.title("Mean for price_data")
plt.show()



# NORMAL DISTRIBUTION FOR PRICE DATA
# Plot between 1000 and 2500 with .001 steps.
x_axis = np.arange(1000, 2500, 0.01)
  
  
plt.plot(x_axis, norm.pdf(x_axis, price_mean, price_stdev))
plt.title("Normal Distribution")
plt.show()

```
